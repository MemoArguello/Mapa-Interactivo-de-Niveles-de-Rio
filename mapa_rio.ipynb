```python
from ipyleaflet import Map, Marker, Icon, basemaps

# Crear el mapa
m = Map(basemap=basemaps.OpenStreetMap.Mapnik, center=[-27.3265, -57.2852], zoom=9)

# Coordenadas de puntos específicos
punto_ayolas= [(-27.4015, -56.8555)]
punto_coratei= [(-27.42691, -56.96995)]
punto_panchito= [(-27.38908, -57.28691)]
punto_cerrito= [(-27.34572, -57.64318)]
punto_itacora= [(-27.23367, -58.20841)]
punto_ppatria= [(-27.25174, -58.55312)]


# Iconos
url_icono_subida = 'https://mapa.comip.ar/images/height_up.png'
icono_subida = Icon(icon_url=url_icono_subida, icon_size=[32, 32])

url_icono_bajada = 'https://mapa.comip.ar/images/height_down.png'
icono_bajada = Icon(icon_url=url_icono_bajada, icon_size=[32, 32])

url_icono_igual = 'https://mapa.comip.ar/images/height_equal.png'
icono_igual = Icon(icon_url=url_icono_igual, icon_size=[32, 32])
```

```python
""" ------------------------------------PASO PATRIA------------------------------------- """
VariacionPasopatria = datos2[1].copy()
Variacion_pp = VariacionPasopatria.loc[5, 'VARIACIÓN DIARIA']
Variacion_pp_numerica = float(Variacion_pp.split()[0])

nivel_pasopatria = datos2[1].copy()
nivel_del_dia_pasopatria = nivel_pasopatria.loc[5, 'NIVEL DEL DÍA']

fecha_pp = datos2[1].copy()
fecha_pasop = fecha_pp.loc[5., 'FECHA']


# Crear el marcador
punto = punto_ppatria[0]
if Variacion_pp_numerica > 0:
    marker = Marker(location=punto, icon=icono_subida, draggable=False)
elif Variacion_pp_numerica < 0:
    marker = Marker(location=punto, icon=icono_bajada, draggable=False)
else:
    marker = Marker(location=punto, icon=icono_igual, draggable=False)

# Crear el cuadro de información (Popup)
popup_content = f"<h2 class='center'>Paso Patria</h2><b>Nivel del día:</b> {nivel_del_dia_pasopatria}" \
                f"<br><b>Variacion:</b> {Variacion_pp}<br><b>Fecha:</b> {fecha_pasop}"
popup = Popup(
    location=punto,
    close_button=True,
    auto_close=False,
    close_on_escape_key=False,
    max_width=800,
    child=HTML(value=popup_content)
)

# Agregar el Popup al marcador
marker.popup = popup

# Agregar el marcador al mapa
m.add_layer(marker)
#Mostrar el mapa
```

```python
""" ------------------------------------ITA CORÁ------------------------------------- """
VariacionItacora = datos2[1].copy()
Variacion_itacora = VariacionItacora.loc[9, 'VARIACIÓN DIARIA']
Variacion_itacora_numerica = float(Variacion_itacora.split()[0])

nivel_itacora = datos2[1].copy()
nivel_del_dia_itacora = nivel_itacora.loc[9, 'NIVEL DEL DÍA']

fecha_ic = datos2[1].copy()
fecha_itac = fecha_ic.loc[9., 'FECHA']

for punto in punto_itacora:
    if Variacion_itacora_numerica > 0:
        marker = Marker(location=punto, icon=icono_subida, draggable=False)
    elif Variacion_itacora_numerica < 0:
        marker = Marker(location=punto, icon=icono_bajada, draggable=False)
    else:
        marker = Marker(location=punto, icon=icono_igual, draggable=False)
        
# Crear el cuadro de información (Popup)
popup_content = f"<h2 class='center'>Ita Corá</h2><b>Nivel del día:</b> {nivel_del_dia_itacora}" \
                f"<br><b>Variacion:</b> {Variacion_itacora}<br><b>Fecha:</b> {fecha_itac}"

popup = Popup(
    location=punto,
    close_button=True,
    auto_close=False,
    close_on_escape_key=False,
    max_width=800,
    child=HTML(value=popup_content)
)

# Agregar el Popup al marcador
marker.popup = popup

# Agregar el marcador al mapa
m.add_layer(marker)
```
